```python
## Creating Auto Process Webscrapping on car_type

def scrape_car_data(html):
    soup = BeautifulSoup(html, "html.parser")
        
    car_models = [a.get_text(strip=True) for a in soup.find_all('a', class_="css-gpjca5 ebw6bgh7")]
    image_sources = [div.find('img')['src'] for div in soup.find_all('div', class_="css-1jqcoh3 e17y54jg1") if div.find('img')]
    car_website = [a['href'] for a in soup.find_all('a', class_="css-gpjca5 ebw6bgh7")]
    
    car_ratings = [div.get_text(strip=True) for div in soup.find_all('div', class_="css-9adkmd ebw6bgh2")]
    
    car_price = [span.get_text(strip=True) for span in soup.find_all('div', class_="css-eee6ji escnsr94")]
    numeric_prices = [int(price.split('$')[1].replace(',', '')) for price in car_price]
    
    car_range = soup.find_all('div', class_="css-1og0o5h escnsr92")
    car_range = [span.get_text(strip=True) for span in soup.find_all('div', class_="css-1og0o5h escnsr92")]

    car_desc = [span.get_text(strip=True) for span in soup.find_all('div', class_="css-1clwbew ek2j0f62")]

       ## Create Dictionary to tabulate pandas dataframe
    return {'image' : image_sources, 
            'website' : car_website,
            'name' : car_models,
            'ratings' : car_ratings,
            'price' : numeric_prices,
            'mileage' : car_range,
            'description' : car_desc}
scrape_car_data(req.text)
```

```python
#Create Car-type listing url
ser_type = pd.Series(['compact', 'mid-size', 'electric/under-60k', 'electric/performance', 'hybrid', 'gas-mileage', 'plug-in-hybrid'])
car_type = pd.DataFrame({'type':ser_type})
car_type
```

```python
#prepare for url car type
def create_cartypec_url(title):
    return 'https://www.caranddriver.com/rankings/best-sedans/' + title
```

```python
car_type['url'] = create_cartypec_url(car_type['type'])
```

```python
urls = car_type['url'].tolist()
urls
```

```python
import time
import random
def perform_random_delay(delay = 3, random_offset =.5):
        time.sleep(delay + random.uniform(0,random_offset))
```

```python
urls = car_type['url'].tolist()
cartype_results  = []

for url in urls:
    # Go to the URL
    req = requests.get(url, headers=headers)
    # If the url is successfully reached, then attempt to scrape the page
    if req.status_code == 200:
        # PERFORM SCRAPING LOGIC HERE
        try:
            curr_url_result = scrape_car_data(req.text)
            curr_url_result['car_url'], curr_url_result['status_code'] = url, req.status_code
            curr_url_result['parsed'] = True
        except: 
            # If the url is not valid, then return an empty dicitonary and denote the status code
            curr_url_result = {'image' : '', 'website': '', 'name' : '', 'ratings' : '', 'price' : '', 'mileage' : '', 'description' : '',
                               'status_code' : req.status_code, 'car_url' : url, 'parsed' : False}
    else: 
        # If the url is not valid, then return an empty dicitonary and denote the status code
        curr_url_result = {'image' : '', 'website': '', 'name' : '', 'ratings' : '', 'price' : '', 'mileage' : '', 'description' : '',
                               'status_code' : req.status_code, 'carc_url' : url, 'parsed' : False}
    # append the result
    cartype_results.append(curr_url_result)
    # perform a random delay between scrapes
    perform_random_delay()
```
